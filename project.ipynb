```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
data=pd.read_csv('C:\\Users\\kailash parihar\\Downloads\\BollywoodMovieDetail.csv',na_values=['na'],parse_dates=['releaseDate'])
data
```

```python
data2=data.dropna()
data2
```

```python
data3=data2.groupby('genre')
data3
```

```python
data4=data3.get_group('Action')
data4
```

```python
data5=data3.get_group('Drama')
data5
```

```python
x=[]
temp1=0
temp2=0
temp3=0
temp4=0
genre=['Drama','Action','Comedy','Romance']
for i in data['genre']:
    if i=='Drama':
        temp1+=1
    if i=='Action':
        temp2+=1
    if i=='Comedy':
        temp3+=1
    if i=='Romance':
        temp4+=1
x.append(temp1)
x.append(temp2)
x.append(temp3)
x.append(temp4)
x
```

```python
plt.bar(genre,x)
plt.title('movies pop')
plt.xlabel('type of movies')
plt.ylabel('number od movies')
plt.show()
```

```python
total=temp1+temp2+temp3+temp4
total
```

```python
Drama=(temp1*100)/total
Action=(temp2*100)/total
Comedy=(temp3*100)/total
Romance=(temp4*100)/total
pl=[Drama,Action,Comedy,Romance]
color=['green','yellow','red','blue']
lab=['Drama','Action','Comedy','Romance']
plt.pie(pl,labels=lab,colors=color)
plt.show()

```
