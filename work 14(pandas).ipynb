```python
titleyear=bd['title']
genre=bd['genre']
tname = []
for z in range(len(titleyear)-1):
    tname.append(titleyear[z])
gname=[]
for z in range(len(genre)):
    gname.append(genre[z])
dic = {
    'title':tname,
    'genre':gname
}
print(len(tname))
new = pd.DataFrame(dic)
new
```
