```python
import bs4
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Machine_learning'
data = requests.get(url)
```

```python
soup = bs4.BeautifulSoup(data.text,'html.parser')

for z in soup.find_all('a'):
    s=str(z.get('href'))
    if s[:4]=='http':
        print(s)
    
```
