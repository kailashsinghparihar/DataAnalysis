## step 4 analysing bookings per day in month of april 2014


```python
dates=[]
y=[]
for z in range(1,31):
    for i in range(len(uber)):
        if uber['Date/Time'][i].day==z:
            dates.append(uber['Date/Time'][i])
    y.append(len(dates))
    dates.clear()
```

```python
y
```

```python
x=list(range(1,31))
x
```

```python
plt.bar(x,y,color='green')
plt.xlabel('dates')
plt.ylabel('no. of bookings')
plt.show()
```

## analysing peak hour

```python
hbookings=[]
temp=0
for z in range(24):
    for i in range(len(uber)):
        if uber['Date/Time'][i].hour==z:
            temp+=1
    hbookings.append(temp)
    temp=0
```

```python
hbookings
```

```python
x=list(range(24))
```

```python
plt.bar(x,hbookings)
plt.title('peak hour analysis')
plt.xlabel('hour')
plt.ylabel('no. of bookings')
plt.show()
```

```python
weekdaybookings=[]
temp=0
list1=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
for z in list1:
    for i in range(len(uber)):
        if uber['Date/Time'][i].weekday_name==z:
            temp+=1
    weekdaybookings.append(temp)
    temp=0
```

```python
weekdaybookings
```

```python
plt.bar(list1,weekdaybookings)
plt.title('week day analysis')
plt.xlabel('weekdays')
plt.ylabel('no. of bookings')
plt.show()
```

```python
plt.plot(list1,weekdaybookings)
plt.title('week day analysis')
plt.xlabel('weekdays')
plt.ylabel('no. of bookings')
plt.show()
```

```python
workingdays=[]
nonworkingdays=[]
total=len(uber)
wd=0
nwd=0
list1=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
for z in list1:
    for i in range(len(uber)):
        if uber['Date/Time'][i].weekday_name=='Monday'or uber['Date/Time'][i].weekday_name=='Tuesday'or uber['Date/Time'][i].weekday_name=='Wednesday'or uber['Date/Time'][i].weekday_name=='Thursday'or uber['Date/Time'][i].weekday_name=='Friday':
            wd+=1
        if uber['Date/Time'][i].weekday_name=='Saturday'or uber['Date/Time'][i].weekday_name=='Sunday':
            nwd+=1
```
